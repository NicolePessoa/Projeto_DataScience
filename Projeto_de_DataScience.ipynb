#**An√°lise de Notas de Filmes Especificos**

```python
notas.query("Filmes==1").Notas.mean()
```

```python
notas.query("Filmes==2").Notas.mean()
```

```python
notas.groupby("Filmes")
```

```python
notas.groupby("Filmes").mean()
```

```python
notas.groupby("Filmes").mean()["Notas"]
```

```python
medias_por_filmes = notas.groupby("Filmes").mean()["Notas"]
medias_por_filmes.head()
```

```python
medias_por_filmes.plot()
```

```python
medias_por_filmes.plot(kind='hist')
```

```python
medias_por_filmes.describe()
```

```python
sns.distplot(medias_por_filmes)
```

```python
sns.distplot(medias_por_filmes, bins=10)
```

```python
sns.displot(medias_por_filmes)
```

#**An√°lise com Matplot**

```python
import matplotlib.pyplot as plt
plt.hist(medias_por_filmes)
plt.title("Histograma das m√©dias dos Filmes")
```

```python
plt.figure(figsize=(6,8))
sns.boxplot(y=medias_por_filmes)
```
